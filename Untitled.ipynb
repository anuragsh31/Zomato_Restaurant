```python
df_country=pd.read_excel('Country-Code.xlsx',)
df_country.head()
```

```python
# for merge two data on a single column
# on is used to match the coloumn on which you combine
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
# To check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
# Pie chart top 3 country uses zomato
plt.pie(country_val[:3], labels=country_name[:3],autopct='%1.2f%%'
       )
```

So, mostly transiction of Zomato is in India. after this US and then UK.

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'No of Rating'})
```

```python
ratings
```

```python
ratings['Rating text'].value_counts()
```

## Conclusion
    Rating     Result
1. 4.5- 4.9 --->   Excellent
2. 4.0- 4.4  --->   Very Good
3. 3.5- 3.9  --->  Good
4. 2.5- 3.4  --->  Avarage
5. 1.8- 2.4  --->  Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='No of Rating', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='No of Rating',hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green'])
```

## Conclusion
1. Not rated count is very high.
2. Maximum no. of rating are between 2.5 to 3.4

```python
#  Count plot
sns.countplot(x='Rating color', data=ratings, palette=['white','red','orange','yellow','green'])
```
